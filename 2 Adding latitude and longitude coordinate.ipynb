```python
#import libraries
import numpy as np
import pandas as pd
import geocoder

#read dataframe from task 1
df= pd.read_csv("Canadadata.csv", index_col=0)

df.head()
```

```python
def get_latlng(postal_code):
    # initialize your variable to None
    lat_lng_coords = None
    # loop until you get the coordinates
    while(lat_lng_coords is None):
        g = geocoder.arcgis('{}'.format(postal_code))
        lat_lng_coords = g.latlng
    return lat_lng_coords
```

```python
#test M5G
get_latlng("M5G")
```

```python
#find coordinates for all postal codes in df 
coords = [get_latlng(postal_code) for postal_code in df["PostalCode"].tolist()]
```

```python
#add columns latitude and longitude to df containing coordinates
df_coords= pd.DataFrame(coords, columns= ["Latitude", "Longitude"])
df["Latitude"]= df_coords["Latitude"]
df["Longitude"]= df_coords["Longitude"]
```
